```python
data['month_year'] = data['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
data['month_year']
```

```python
df_trend = data.groupby('month_year').sum().sales.reset_index()
df_trend
```

```python
import matplotlib as plt
```

```python
plt.pyplot.figure(figsize=(15,6))
ax = plt.pyplot.axes()
ax.set_facecolor("yellow")
plt.pyplot.xticks(rotation='vertical',size=8)
plt.pyplot.plot(df_trend['month_year'],df_trend['sales'],color='red')
plt.pyplot.title("Overall Sales Trend")
plt.pyplot.show()
```
